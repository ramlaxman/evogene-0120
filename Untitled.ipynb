```python
# ask the user for two pieces of information (in two variables)
#   - name
#   - city
# then print one of four responses:
# (1) both are the same as me
# (2) city is the same
# (3) name is the same
# (4) neither is the same 


name = input("Enter your name: ")
city = input("Enter your city: ")

if name == 'Reuven' and city == 'Modiin':
    print("Hey!  You must be me!")
    
elif name == 'Reuven':
    print("Ah, a Reuven from another city!")
    
elif city == 'Modiin':
    print("From my city, but not my name?  OK!")
    
else:
    print("I don't want anything to do with you.")
```
