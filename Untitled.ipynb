```python
# ask the user for two pieces of information (in two variables)
#   - name
#   - city
# then print one of four responses:
# (1) both are the same as me
# (2) city is the same
# (3) name is the same
# (4) neither is the same 


name = input("Enter your name: ")
city = input("Enter your city: ")

if name == 'Reuven' and city == 'Modiin':
    print("Hey!  You must be me!")
    
elif name == 'Reuven':
    print("Ah, a Reuven from another city!")
    
elif city == 'Modiin':
    print("From my city, but not my name?  OK!")
    
else:
    print("I don't want anything to do with you.")
```

```python
# (1) generate a random number from 0-100
import random
number = random.randint(0, 100)

# (2) print the random number
print(number)

# (3) ask the user to guess
guess = int(input("Enter a guess: "))

# (4) print that it's right, too high, or too low
if guess == number:
    print("You got it!")
elif guess < number:
    print("Too low!")
elif guess > number:
    print("Too high!")
else:
    print("You should never see this")
```
