```python
# Pig Latin

# (1) If first letter is a, e, i, o, or u -- add "way"
# air -> airway
# elephant -> elephantway
# octopus -> octopusway

# (2) Otherwise, move the first letter to the end, and add "ay"
# computer -> omputercay
# telephone -> elephonetay
# papaya -> apayapay

# https://github.com/reuven/evogene-0120

word = input("Enter a word: ")

if word[0] in 'aeiou':
    print(word + 'way')
else:
    print(word[1:] + word[0] + 'ay')
```
