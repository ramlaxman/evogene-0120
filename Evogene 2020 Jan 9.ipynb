```python
menu = {'sandwich': 30, 'soup': 25, 'tea': 10}
total = 0

while True:
    order = input("Order: ").strip()
    if not order:
        break
    elif order in menu:
        price = menu[order]
        total += price
        print(f'{order} costs {price}, total now {total}')
    else:
        print(f'We do not sell {order} here!')

print(f'Your total is {total}')
```

```python
d = {'a':1, 'b':2, 'c':3}

while True:
    k = input("Enter a key: ").strip()
    if not k:
        break
    elif k in d:
        print(f'd[{k}] = {d[k]}')
    else:
        print(f'{k} not in d')
```

```python
d.get('a')   # if 'a' in d, d['a']
```

```python
d = {'a':1, 'b':2, 'c':3}

while True:
    k = input("Enter a key: ").strip()
    if not k:
        break
    else:
        print(f'd[{k}] = {d.get(k, "Not defined")}')
```
