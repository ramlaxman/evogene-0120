```python
menu = {'sandwich': 30, 'soup': 25, 'tea': 10}
total = 0

while True:
    order = input("Order: ").strip()
    if not order:
        break
    elif order in menu:
        price = menu[order]
        total += price
        print(f'{order} costs {price}, total now {total}')
    else:
        print(f'We do not sell {order} here!')

print(f'Your total is {total}')
```

```python
d = {'a':1, 'b':2, 'c':3}

while True:
    k = input("Enter a key: ").strip()
    if not k:
        break
    elif k in d:
        print(f'd[{k}] = {d[k]}')
    else:
        print(f'{k} not in d')
```

```python
# Numbers: 10 20 30
# Total is 60 

total = 0

s = input("Enter numbers: ")

for one_number in s.split():
    if one_number.isdigit():
        print(one_number)
        total += int(one_number)
    else:
        print(f'Ignoring {one_number}')
    
print(f'Total = {total}')
```

```python
d1 = {'a':1, 'b':2,'c':3, 'd':4}
d2 = {'a':1, 'b':20, 'c':30}

for one_key in d1.keys() | d2.keys():
    if d1.get(one_key) != d2.get(one_key):
        print(f'{one_key} different')
```
